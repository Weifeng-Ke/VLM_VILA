```python
# Imports
import importlib.util
import json
import os
import re
import cv2
from pydantic import BaseModel
from termcolor import colored
import llava
from llava import conversation as clib
from llava.media import Image, Video
from llava.model.configuration_llava import JsonSchemaResponseFormat, ResponseFormat

# --- configuration ---
model_path = "Efficient-Large-Model/VILA1.5-3b"
conv_mode = "vicuna_v1"
text_prompt = "Please describe the image"
image_path = "/workspace/VILA/demo_images/demo_img.png"

# --- load model ---
print("Loading ViLA model...")
model = llava.load(model_path, model_base=None)

# --- set conversation template (same as --conv-mode) ---
clib.default_conversation = clib.conv_templates[conv_mode].copy()

# --- prepare multimodal prompt ---
media = Image(image_path)
prompt = [media, text_prompt]

# --- run inference ---
print("Generating response...")
response = model.generate_content(prompt)

# --- display result ---
from termcolor import colored
print(colored(response, "cyan", attrs=["bold"]))
```
